```python
from scipy.cluster.hierarchy import linkage
```

```python
from sklearn.datasets import load_iris

iris = load_iris()

data = iris['data']

data[0:5]
```

```python
hclusters = linkage(data, 'ward')
```

```python
hclusters
```

```python
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist

# shows distance between observations

c, coph_dists = cophenet(hclusters, pdist(data))
```

```python
c
```

```python
coph_dists
```

```python
from scipy.cluster.hierarchy import dendrogram
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.title('Dendrogram')
plt.xlabel('Index Numbers')
plt.ylabel('Distance')
dendrogram(
    hclusters,
    leaf_rotation=90.,  
    leaf_font_size=8.,
)
plt.show()
```

```python
plt.figure(figsize=(20,16))
plt.title('Dendrogram')
plt.xlabel('Index Numbers')
plt.ylabel('Distance')
dendrogram(
    hclusters,
    leaf_rotation=90.,  
    leaf_font_size=8.,
)
plt.show()
```

```python
hclusters[0:2]
```

```python
from scipy.cluster.hierarchy import fcluster

# shows which cluster each point belongs to

max_dist = 10
hclusters = fcluster(hclusters, max_dist, criterion='distance')
hclusters
```

```python
# takes dict or list and returns unique values

set(hclusters)
```

```python
plt.scatter(data[:,0], data[:,2], c=hclusters, cmap='viridis')
plt.show
```

```python
import pandas as pd
import os
%pwd
```

```python
wines = pd.read_csv('../downloads/wine (1).csv')
```

```python
wines.head()
```

```python
quality = wines['quality']
wines.drop('quality', axis=1, inplace=True)
wines.head()
```

```python
wines.shape
```

```python
from sklearn.preprocessing import StandardScaler

wines_standardized = StandardScaler().fit_transform(wines.values)
```

```python
wine_clusters = linkage(wines_standardized, 'ward')
```

```python
wine_clusters[0]
```

```python
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist

# shows distance between observations

c, coph_dists = cophenet(wine_clusters, pdist(wines_standardized))
```

```python
c
```

```python
plt.figure(figsize=(20,16))
plt.title('Dendrogram for Wine')
plt.xlabel('Index Numbers')
plt.ylabel('Distance')
dendrogram(
    wine_clusters,
    leaf_rotation=90.,  
    leaf_font_size=8.,
)
plt.show()
```

```python
# we know there are 6 different wine qualities
# so we should look for a distance that will give us 6 clusters

quality.value_counts()
```

```python
max_dist = 39
clusters = fcluster(wine_clusters, max_dist, criterion='distance')
set(clusters)
```

```python
import seaborn as sns

sns.pairplot(wines)
```

```python
plt.scatter(wines['alcohol'], wines['pH'], c=clusters, cmap='viridis')
plt.show
```

```python
# quality isnt showing a distinct cluster
# there are other things that make these wines similar with a good rating

plt.scatter(wines['alcohol'], wines['pH'], c=quality, cmap='viridis')
plt.show
```

```python
x = df[[col for col in df.columns if col != 'Class']]
x.head()
```

```python
napolean = StandardScaler().fit_transform(x)
napolean[0:5]
```

```python
napolean.shape
```

```python
# create covariance matrix how often does one voting pattern predict the other

covariance_matrix = np.cov(napolean.T)
covariance_matrix[0:5]
```

```python
covariance_matrix.shape
```

```python
# eigen 

eigenvalues, eigenvactors = np.linalg.eig(covariance_matrix)
```

```python
# eqv to transforming it 7.10 times, etc
eigenvalues
```

```python
eigenvalues.shape
```

```python
eigenvactors[0:3]
```

```python
eigenvactors.shape
```

```python
# find principal components

eigenpairs = zip(eigenvalues, eigenvactors)
eigenpairs
```

```python
for i in sorted(eigenpairs, reverse=True):
    print i
```

```python
# if we were to apply this to the whole 16 X 16 square it would scale it up this sum

eigansums = sum(eigenvalues)
eigansums
```

```python
for eig in sorted(eigenpairs, reverse=True):
    print eig[0] / eigansums, eig[0]
    
# first transformation will give us 44% second 9%
```

```python
np.cumsum([(eig / eigansums) * 100 for eig in sorted(eigenvalues, reverse=True)])
```

```python
Pmatrix = np.hstack(fit_me)
Pmatrix
```

```python
pca_data = napolean.dot(Pmatrix)
pca_data
```

```python
pca = PCA(n_components=8)
pca_x = pca.fit_transform(napolean)
pca_x[0:5]
```

```python
from sklearn.cluster import KMeans

regular_model = KMeans(2)

pca_model = KMeans(2)
```

```python
def how_many_ones(x):
    ones = [y for y in x if y ==1]
    return len(ones) / float(len(x))
```
